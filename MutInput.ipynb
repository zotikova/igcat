```python
import random

aa_alphabet = ['A', 'R', 'N', 'D']
mut_percent = 10 #in %

with open('/home/sola/py_projects/igcat/seq_data/inputs/HEAVY_id_comb.fasta') as H_ini:
    H_ini = H_ini.readlines()
    for i in range(len(H_ini)):
        if H_ini[i][0] == '>':
            str_len = len(H_ini[i+1])+len(H_ini[i+2])
            str_aa = H_ini[i+1]+H_ini[i+2]
            N_aa_mut = round(str_len*mut_percent/100)
            for N in range(N_aa_mut):
                ind_aa_mut = random.choice(range(str_len))
                aa_mut = str_aa[ind_aa_mut]
                aa_subst = random.choice(aa_alphabet)
                #print(str_aa, '\n')
                if aa_subst != aa_mut:
                    str_aa = str_aa[0:ind_aa_mut] + aa_subst + str_aa[ind_aa_mut+1:]
            print(str_aa)
```
