## Prepare IgBLAST output in consonance with IgCAT output

```python
def set_var(value):
    if value and value.strip():
        return value
    else:
        return "N/A N/A N/A N/A N/A N/A"

# for .blasted files

def igblast_prep(output):
    line_steck = []
    with open(output, 'r') as file:
        lines = file.readlines()
        for line in range(len(lines)):
            if lines[line][:7] == 'Query= ' and set_var(lines[line+14]).split()[1] == 'summary' and set_var(lines[line+15]).split()[0] == 'FR1' and set_var(lines[line+16]).split()[0] != 'Total':
                pdbid = lines[line][7:-1]
                #print(pdbid)
                fr1_beg = set_var(lines[line+15]).split()[1]
                fr1_end = set_var(lines[line+15]).split()[2]
                cdr1_beg = set_var(lines[line+16]).split()[1]
                cdr1_end = set_var(lines[line+16]).split()[2]
                fr2_beg = set_var(lines[line+17]).split()[1]
                fr2_end = set_var(lines[line+17]).split()[2]
                cdr2_beg = set_var(lines[line+18]).split()[1]
                cdr2_end = set_var(lines[line+18]).split()[2]
                fr3_beg = set_var(lines[line+19]).split()[1]
                fr3_end = set_var(lines[line+19]).split()[2]
                cdr3_beg = set_var(lines[line+20]).split()[4]
                cdr3_end = set_var(lines[line+20]).split()[5]
                #print(fr1_beg)
                line_steck.append([pdbid, fr1_beg, fr1_end, cdr1_beg, cdr1_end, fr2_beg, fr2_end, cdr2_beg, cdr2_end, fr3_beg, fr3_end, cdr3_beg, cdr3_end, '\n'])
    return line_steck
                
try_igblast = igblast_prep('/home/sola/py_projects/igcat/seq_data/raw_outputs/igblast_comb_heavy.blasted')

# writedown kabat file

with open('/home/sola/py_projects/igcat/seq_data/prep_outputs/igblast_comb_heavy.kabat', 'w') as file:
        for line in try_igblast:
            line2 = [x if x != 'N/A' else '0' for x in line]
            line3 = [x if x != '(V' else '0' for x in line2]
            line4 = [x if x != 'gene' else '0' for x in line3]
            line1 = '	'.join(line4)
            file.write(line1)
```
